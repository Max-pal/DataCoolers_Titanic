```python
train = pd.read_csv("input/titanic/train.csv")
test = pd.read_csv("input/titanic/test.csv")
```

```python
train.describe()
test.describe()
```

```python
train.head()
test.head()
```

```python
train=train.drop(columns=["Cabin", "Ticket"])
test=test.drop(columns=["Cabin", "Ticket"])
```

```python
train.fillna(train.mean(), inplace=True)
test.fillna(test.mean(), inplace=True)
```

```python
train.head()
test.head()
```

```python
train.describe()
test.describe()
```

```python
Onehot = pd.get_dummies(train['Embarked'])

train = pd.concat([train, Onehot], axis=1)
```

```python
Onehot = pd.get_dummies(test['Embarked'])

test = pd.concat([test, Onehot], axis=1)
```

```python
train.head()
```

```python
train.shape
```

```python
Onehot = pd.get_dummies(train['Sex'])
train = pd.concat([train, Onehot], axis=1)
```

```python
Onehot = pd.get_dummies(test['Sex'])
test = pd.concat([test, Onehot], axis=1)
```

```python
train = train.drop(columns=["Embarked", "Sex", "Name"])
train.head()
```

```python
test = test.drop(columns=["Embarked", "Sex", "Name"])
test.head()
```

```python

y = train.Survived
```

```python
train.drop(['Survived'], axis=1, inplace=True)
```

```python
train.head()
```

```python
X_train_full, X_valid_full, y_train, y_valid = train_test_split(train, y, train_size=0.8, test_size=0.2,
                                                                random_state=0)
```

```python
list(np.arange(0.05, 0.7, 0.05))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import ShuffleSplit

param_grid = [
  {'n_estimators': list(np.arange(80,200,20)), 'max_depth': [4,5,6,7,8], 'min_samples_leaf': [2,3,4]}
 ]

# , cv=6
my_model_1 = GridSearchCV(RandomForestClassifier(), param_grid, cv=ShuffleSplit(test_size=0.20, n_splits=1),
                       scoring='%s_macro' % 'precision')
my_model_1.fit(X_train_full, y_train)

# my_model_1 = XGBClassifier(random_state=1, learning_rate=0.5, max_depth=3)

# my_model_1.fit(X_train_full, y_train)
```
